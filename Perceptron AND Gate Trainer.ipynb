```python
import matplotlib.pyplot as plt
import random
import math
```

```python
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
Y = [0, 0, 0, 1]  # AND operation
# Y = [0, 1, 1, 1] # OR operation
# Y = [0, 1, 1, 0] # XOR operation
```

```python
plt.figure()
for i in range(len(Y)):
    if Y[i] == 0:
        plt.plot(X1[i], X2[i], "bo")
    else:
        plt.plot(X1[i], X2[i], "ro")
plt.show()
```

```python
def train_perceptron(lr):
    W = [random.uniform(-1, 1), random.uniform(-1, 1)]  # Random initialization between -1 and 1
    b = random.uniform(-1, 1)
    print(f"Initial Weights: {W}, Bias: {b}")

    epoch = 0
    count = 0

    while count < 4:
        total_error = 0
        for i in range(len(Y)):
            # Compute weighted sum
            ws = W[0] * X1[i] + W[1] * X2[i] + b
            pred = 0
            if ws > 0.5:
                pred = 1

            # Compute error
            error = Y[i] - pred
            if error != 0:
                count = 0
                total_error += math.sqrt(error**2)
                # Update weights and bias
                W[0] += lr * error * X1[i]
                W[1] += lr * error * X2[i]
                b += lr * error
            else:
                count += 1

        epoch += 1
        plt.figure()
        for i in range(len(Y)):
            if Y[i] == 0:
                plt.plot(X1[i], X2[i], "bo")
            else:
                plt.plot(X1[i], X2[i], "ro")

        # Plot decision boundary
        xx = plt.xlim()
        slope = -(W[0] / W[1])
        yint = (0.5 - b) / W[1]
        yy = [slope * x + yint for x in xx]
        plt.plot(xx, yy)

        plt.xlim([-1, 2])
        plt.ylim([-1, 2])
        plt.title(f"Learning Rate: {lr}, Epoch: {epoch}")
        plt.show()

    return W, b, epoch
```

```python
print("Training with lr = 0.01")
W1, b1, epoch1 = train_perceptron(0.01)
print(f"Final Weights: {W1}, Bias: {b1}, Epochs: {epoch1}")
```

```python
print("Training with lr = 0.5")
W2, b2, epoch2 = train_perceptron(0.5)
print(f"Final Weights: {W2}, Bias: {b2}, Epochs: {epoch2}")
```
